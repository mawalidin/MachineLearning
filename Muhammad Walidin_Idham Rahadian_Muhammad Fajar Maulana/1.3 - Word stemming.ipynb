```python
import pandas as pd
import numpy as np
import string
import re
from collections import Counter
```

```python
raw = pd.read_csv( "data/data/1.2_data_review_cleaner_pure.csv")
```

```python
raw['review'] = raw['review'].apply(lambda x: x.replace(","," "))
raw
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
raw['review'] = raw['review'].apply(lambda x: stemmer.stem(x).replace(" ",","))
raw
```

```python
def avg_word(words):
  return (sum(len(word) for word in words)/len(words))

raw['avg_word'] = raw['review'].apply(lambda x: avg_word(x.split(",")))
```
