---
 - Machine Learning (ML) merupakan sebuah aplikasi dari kecerdasan buatan. Proses dalam ML ialah kita memerintahkan mesin untuk belajar dan melakukan improvisasi dari pengalaman (data), kemudian membuat sebuah keputusan atau prediksi di masa depan.
 - Data predictor juga disebut sebagai features dan dalam dunia statistik sebagai independent variables. Sedangkan data target dikenal juga sebagai label dan dalam dunia statistik sebagai dependet variables.
 - Supervised Learning ialah metode belajar mesin dimana Algoritma belajar menggunakan labeled data pada input maupun output. Sedangkan Unsupervised Learning ialah metode belajar mesin dimana Algoritma belajar menggunakan no labeled data pada input maupun output.
 - Klasifikasi dan Regresi memiliki perbedaan pada output. Klasifikasi menoutputkan label dan Regresi menoutputkan kuantiti yang kontinu.
 - Hyperparameter merupakan konfigurasi variabel terhadap model. Hyperparameter didefinisikan di awal sebelum model dilatih dengan data. Hyperparameter mempengaruhi kecepatan dan akurasi proses belajar model. Adapun Parameter ialah variabel yang nilainya diestimasikan dari dataset, nilai yang dipelajari dari data selama training.

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# model_tr1 = DecisionTreeRegressor(max_depth=5)
# model_tr2 = DecisionTreeRegressor(max_depth=2)

# x_matriks = x[:, np.newaxis]
# model_tr1.fit(x_matriks,y)
# model_tr2.fit(x_matriks,y)

# x_new = x_new[:, np.newaxis]
# y1 = model_tr1.predict(x_matriks)
# y2 = model_tr2.predict(x_new)

# fig, ax = plt.subplots(figsize=(12,8))
# ax.scatter(x,y,label='Data Asli')
# ax.scatter(x_matriks,y,c='yellow')
# ax.plot(x_new,y2,c='green',label='Prediksi Terhadap Data Baru')
# ax.legend(loc='upper right',bbox_to_anchor=[1.2,1.2])
model_lr = LinearRegression(fit_intercept=True)

x_matriks = x[:,np.newaxis]

model_lr.fit(x_matriks,y)
```

```python
x_new = x_new[:, np.newaxis]
y_test = model_lr.predict(x_new)
y_test
```

```python
y_train = model_lr.predict(x_matriks)

plt.scatter(x,y, label='Data Asli',c='green')
plt.plot(x, y_train,label='Prediksi Terhadap Data Training')
plt.plot(x_new, y_test, label='Prediksi terhadap data baru')
plt.legend(loc='upper right',bbox_to_anchor=[1.3,1.3])
plt.show()
```
